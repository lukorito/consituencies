```python
import tabula as tb
import pandas as pd
import re
import numpy as np
import json
```

```python
file = 'polling.pdf'
data = tb.read_pdf(file, pages = 'all')
```

```python
df = pd.DataFrame(np.concatenate(data))
```

```python
# Drop NAN
df.dropna(inplace=True)
```

```python
# convert row to column header
df.columns = df.iloc[0]
df = df.drop(0)
df = df.reset_index(drop=True)
```

```python
df
```

```python
all_data = []
```

```python
for index, row in df.iterrows():
    name = row["County Name"]
    code = row["County\rCode"]
    all_data.append({
        name, code
    })
```
